```python
schema = define_schema()
data = spark.read.option("mergeSchema", "true").schema(schema).parquet(streaming_data_path)
```

```python
from pyspark.sql.functions import isnan, when, count, col
```

```python
not_date = list(set(data.columns) - {'hour', 'date'})
```

```python
nulls = data.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in data[not_date].columns])
```

```python
nulls.toPandas().transpose().sort_index()
```
